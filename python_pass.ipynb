## Implement the isolate function that:
## 1- Takes students dictionary where each student’s value is his class.
## 2- Returns a dicitionary of classes as keys where each class has a value of list of students in that class, in any order.

```python
def sort_dict_by_value(d, reverse = False):
  return dict(sorted(d.items(), key = lambda x: x[1], reverse = reverse))

print("Original dictionary elements:")
students = {
    'Ahmed': 'Class 1',
    'Mustafa': 'Class 3',
    'Ali': 'Class 2',
    'Sara': 'Class 1',
    'Zainab': 'Class 1',
    'Zain': 'Class 2',     
}

print(students)
print("\nSort (ascending) the said dictionary elements by value:")
print(sort_dict_by_value(students))
```
